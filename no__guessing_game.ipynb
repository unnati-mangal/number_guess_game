```python
import random
import time
import os

LEADERBOARD_FILE = "leaderboard.txt"




def load_leaderboard():
    if not os.path.exists(LEADERBOARD_FILE):
        return []

    scores = []
    with open(LEADERBOARD_FILE, "r") as file:
        for line in file:
            line = line.strip()
            if line:
                name, score = line.split(",")
                scores.append((name, int(score)))
    return scores


def save_leaderboard(scores):
    with open(LEADERBOARD_FILE, "w") as file:
        for name, score in scores:
            file.write(f"{name},{score}\n")


def update_leaderboard(player_name, player_score):
    scores = load_leaderboard()
    scores.append((player_name, player_score))
    scores.sort(key=lambda x: x[1], reverse=True)
    scores = scores[:5]  # keep top 5
    save_leaderboard(scores)


def show_leaderboard():
    scores = load_leaderboard()
    print("\nğŸ† TOP 5 LEADERBOARD ğŸ†")

    if not scores:
        print("No scores yet. Be the first champion ğŸ˜„")
        return

    for i, (name, score) in enumerate(scores, start=1):
        print(f"{i}. {name} - {score} points")

def choose_difficulty():
    print("\nğŸš Choose Difficulty:")
    print("1. Easy   (1-50)   Attempts: 10")
    print("2. Medium (1-100)  Attempts: 7")
    print("3. Hard   (1-500)  Attempts: 8")

    while True:
        choice = input("Enter choice (1/2/3): ").strip()

        if choice == "1":
            return 1, 50, 10, 1  # (start, end, attempts, multiplier)
        elif choice == "2":
            return 1, 100, 7, 2
        elif choice == "3":
            return 1, 500, 8, 3
        else:
            print("âš ï¸ Invalid choice. Please enter 1, 2, or 3.")


def calculate_score(max_attempts, attempts_used, time_taken, multiplier):
    # Base score depends on attempts left
    attempts_score = (max_attempts - attempts_used + 1) * 10

    # Time bonus (faster = more score)
    if time_taken <= 15:
        time_bonus = 30
    elif time_taken <= 30:
        time_bonus = 20
    elif time_taken <= 60:
        time_bonus = 10
    else:
        time_bonus = 0

    final_score = (attempts_score + time_bonus) * multiplier
    return max(0, final_score)


def play_game(player_name):
    start, end, max_attempts, multiplier = choose_difficulty()

    secret = random.randint(start, end)

    low_range = start
    high_range = end

    attempts = 0
    guesses = []

    print(f"\nğŸ® Welcome {player_name}!")
    print(f"Guess the number between {start} and {end}")
    print(f"You have {max_attempts} attempts.\n")

    start_time = time.time()

    while attempts < max_attempts:
        print(f"ğŸ” Current Range: {low_range} to {high_range}")
        print(f"Attempts left: {max_attempts - attempts}")

        try:
            guess = int(input("Enter your guess: "))
        except ValueError:
            print("âš ï¸ Enter a valid number!\n")
            continue

        if guess < low_range or guess > high_range:
            print("ğŸš« Guess within the current range!\n")
            continue

        attempts += 1
        guesses.append(guess)

        if guess == secret:
            end_time = time.time()
            time_taken = int(end_time - start_time)

            score = calculate_score(max_attempts, attempts, time_taken, multiplier)

            print("\nğŸ‰ YOU WON!")
            print(f"âœ… Correct Number: {secret}")
            print(f"ğŸ¯ Attempts Used: {attempts}")
            print(f"â± Time Taken: {time_taken} seconds")
            print(f"ğŸ† Your Score: {score}")
            print(f"ğŸ“œ Your Guesses: {guesses}")

            update_leaderboard(player_name, score)
            return True, score

        # Update range based on guess
        if guess < secret:
            low_range = guess + 1
        else:
            high_range = guess - 1

        print("âŒ Wrong guess! Range updated.\n")

    # If user fails
    print("\nâŒ GAME OVER!")
    print(f"ğŸ¯ The correct number was: {secret}")
    print(f"ğŸ“œ Your Guesses: {guesses}")
    return False, 0


def start():
    print("======================================")
    print("ğŸ² ADVANCED NUMBER GUESSING GAME ğŸ²")
    print("======================================")

    player_name = input("Enter your name:     if player_name == "":
        player_name = "Player"

    games_played = 0
    wins = 0
    total_score = 0

    while True:
        games_played += 1
        won, score = play_game(player_name)

        if won:
            wins += 1
            total_score += score

        print("\nğŸ“Š GAME STATS")
        print("Games Played:", games_played)
        print("Wins:", wins)
        print("Losses:", games_played - wins)
        print("Total Score:", total_score)

        show_leaderboard()

        choice = input("\nDo you want to play again? (yes/no): ").lower().strip()
        if choice != "yes":
            print("\nğŸ‘‹ Thanks for playing! See you next time.")
            break


start()
```
