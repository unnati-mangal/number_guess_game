```python

movie_1 = "Dhurandhar"
price_1 = 200
age_limit_1 = 18

movie_2 = "PK"
price_2 = 250
age_limit_2 = 0

movie_3 = "Chennai Express"
price_3 = 200
age_limit_3 = 0

movie_4 = "People we meet on vacation"
price_4 = 300
age_limit_4 = 18


print("Available Movies")
print("1.", movie_1, "-", price_1, "rs", "| Age:", age_limit_1, "+")
print("2.", movie_2, "-", price_2, "rs", "| Age: All")
print("3.", movie_3, "-", price_3, "rs", "| Age: All")
print("4.", movie_4, "-", price_4, "rs", "| Age:", age_limit_4, "+")

# Customer Details
NAME = input("\nEnter your name: ")
AGE = int(input("Enter your age: "))

movie_name = input("\nEnter the movie name: ")

# Set price + age limit
if movie_name == movie_1:
    price = price_1
    age_limit = age_limit_1

elif movie_name == movie_2:
    price = price_2
    age_limit = age_limit_2

elif movie_name == movie_3:
    price = price_3
    age_limit = age_limit_3

elif movie_name == movie_4:
    price = price_4
    age_limit = age_limit_4

else:
    print("Movie not available!")
    exit()

# Age check (movie wise)
if AGE < age_limit:
    print("\nSorry! This movie is", age_limit, "+")
    print("You are not allowed to watch this movie.")
    exit()

print("\nMovie selected:", movie_name)
print("Ticket price:", price, "rs")

# Date input
date = input("\nEnter date (example: 10-02-2026): ")

# Slot selection
print("\nAvailable Slots")
print("1. Morning (10:00 AM)")
print("2. Afternoon (2:00 PM)")
print("3. Evening (6:00 PM)")
print("4. Night (9:00 PM)")

slot_choice = int(input("\nChoose slot number: "))

if slot_choice == 1:
    slot = "Morning (10:00 AM)"
elif slot_choice == 2:
    slot = "Afternoon (2:00 PM)"
elif slot_choice == 3:
    slot = "Evening (6:00 PM)"
elif slot_choice == 4:
    slot = "Night (9:00 PM)"
else:
    print("Invalid slot choice!")
    exit()

tickets = int(input("\nEnter total tickets to be booked: "))

total_amount = tickets * price
print("\n==============================")
print("BOOKING SUMMARY")
print("==============================")
print("Name:", NAME)
print("Age:", AGE)
print("Movie:", movie_name)
print("Date:", date)
print("Slot:", slot)
print("Ticket Price:", price, "rs")
print("Total Tickets:", tickets)
print("Total Amount:", total_amount, "rs")
print("==============================")
print("\nâœ… Booking Confirmed!")
```
